```python
from databricks.labs.gbx.sample import get_volumes_path, run_essential_bundle
from pathlib import Path

CATALOG = "main"
SCHEMA = "default"
VOLUME = "geobrix_samples"
SAMPLE_DATA_PATH = get_volumes_path(CATALOG, SCHEMA, VOLUME)
print(f"SAMPLE_DATA_PATH = {SAMPLE_DATA_PATH}")
```

```python
Path(SAMPLE_DATA_PATH).mkdir(parents=True, exist_ok=True)
result = run_essential_bundle(SAMPLE_DATA_PATH)
print(f"Files: {result['file_count']}, Total: {result['total_size_mb']:.1f} MB")
```
